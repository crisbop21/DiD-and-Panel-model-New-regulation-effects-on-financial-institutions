```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('DiD_data.csv')
df2 = df.dropna()
df2 = df2.reset_index(drop=True)
```

```python
df2
```

# Panel B Test 1

```python
PanelB_1 = smf.ols(formula='bhc_avgtradingratio ~ after_DFA_1*treat_3_b_avg + C(rssd9001) + C(rssd9999) + dep_roa1', data=df2).fit()
print(PanelB_1.summary())
```
