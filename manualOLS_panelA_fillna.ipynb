```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline

import statsmodels.formula.api as sm
```

```python
df = pd.read_csv('DiD_data.csv')
df.head()
```

```python
df.info()
```

## Panel A

```python
affect = pd.read_csv('df_pre09_BHC_Affect_var.csv')
affect.head()
```

```python
affect.info()
```

```python
affect['Affect'].fillna(value=0, inplace=True)
```

```python
df_affect = df.merge(affect, left_on = 'rssd9001', right_on = 'rssd9001', how = 'left')
```

```python
df_affect.info()
```

```python
#df_affect.dropna(inplace=True)
```

```python
df_affect.info()
```

```python
panelA_1 = sm.ols(formula='bhc_avgtradingratio ~ after_DFA_1', data=df_affect).fit()
print(panelA_1.summary())
```

```python
panelA_2 = sm.ols(formula='bhc_avgtradingratio ~ after_DFA_1 + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter', data=df_affect).fit()
print(panelA_2.summary())
```

```python
panelA_3 = sm.ols(formula='bhc_avgtradingratio ~ after_DFA_1*Affect + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter', data=df_affect).fit()
print(panelA_3.summary())
```

```python
df_affect.dropna(inplace=True)
```

```python
panelA_4 = sm.ols(formula='bhc_avgtradingratio ~ after_DFA_1*Affect + C(rssd9001) + C(rssd9999) + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter', data=df_affect).fit()
print(panelA_4.summary())
```
